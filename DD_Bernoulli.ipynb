```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from scipy.stats import bernoulli
p = 0.3
data = bernoulli.rvs(p, size = 100)
len(data)
```

```python
mend, var, skew, kurt = bernoulli.stats(p, moments= 'mvsk')
mend, var, skew, kurt
```

```python
import seaborn as sns
ax = sns.distplot(data, bins = 30, kde = False, color='green')
ax.set(xlabel ='Distribucion Bernoulli', ylabel ='frecuencia')
ax.axvline(x=mend, linestyle='--',label='Media', c='r')

ax.legend()
```
