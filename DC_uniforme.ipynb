```python
import matplotlib.pyplot as plt 
%matplotlib inline

from scipy.stats import uniform
```

```python
data = uniform.rvs(size =100000000)
len(data)
```

```python
mead, var, skew, kurt = uniform.stats(moments='mvsk')
mead, var, skew, kurt
```

```python
import seaborn as sns 
ax = sns.distplot(data, bins=500,kde=False, color='green')
ax.set(xlabel='tiempo', ylabel='frecuencia')
ax.axvline(x = mead,linestyle='--',label='Media',c='r')
ax.axvline(x = var,linestyle='-.',label='Varianza',c='c')
ax.set_title('Distribucin uniforme 0 -1')
ax.legend()
```
