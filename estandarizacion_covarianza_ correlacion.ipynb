```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('iris-data.csv', index_col=0)
```

```python
df.columns
```

```python
df.tipo_flor.value_counts()
```

```python
y = df['lar.petalo']
```

```python
fig, ax = plt.subplots()    
ax.set(xlabel='X',ylabel='Y')
ax.set_title('Variable original')
ax.hist(y, bins=30)
ax.axvline(x=np.mean(y), c='g', linestyle='--', label='Media')
ax.axvline(x=np.mean(y)+np.std(y), c='r', linestyle='--', label='Desviacion estandar')
ax.legend()
```

```python
fig, ax = plt.subplots()    
ax.set(xlabel='X',ylabel='Y')
ax.set_title('Estandarizacion de la variable') #Es centrar sobre el eje 0
ax.hist(y - np.mean(y), bins=30)
ax.axvline(x=np.mean(y - np.mean(y)), c='g', linestyle='--', label='Media')
ax.axvline(x=np.mean(y)+np.std(y), c='r', linestyle='--', label='Desviacion estandar')
ax.legend()
```

```python
fig, ax = plt.subplots()    
ax.set(xlabel='X',ylabel='Y')
ax.set_title('Reduccion de la variable') #Es centrar sobre el eje 0
ax.hist(y - np.mean(y), bins=30)
ax.axvline(x=np.mean(y - np.mean(y)), c='g', linestyle='--', label='Media')
ax.axvline(x=np.mean(y)+np.std(y), c='r', linestyle='--', label='Desviacion estandar')
ax.legend()
```
