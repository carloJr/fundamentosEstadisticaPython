```python
import matplotlib.pyplot as plt 
from IPython.core.display import Image
import seaborn as sns 

%matplotlib inline
```

```python
from scipy.stats import t

data1 = t.rvs(100, size=1000000)
data2 = t.rvs( 4, size=1000000)

ax=sns.distplot(data1, bins=500, kde=False, color='b')
ax=sns.distplot(data2, bins=500, kde=False, color='r')
ax.set(xlabel='Grados de libertad',ylabel='Frecuencia')
```

```python
from scipy.stats import chi2 # hi cuadrado que permite calcular la forma para calculos de probabilidad para varianzas con "chi2"

data1 = chi2.rvs( 5, size=1000000)
data2 = chi2.rvs(15, size=1000000)

ax=sns.distplot(data1, bins=500, kde=False, color='b')
ax=sns.distplot(data2, bins=500, kde=False, color='r')
ax.set(xlabel='Grados de libertad',ylabel='Frecuencia')
```

```python
from scipy.stats import f # Nos permite aproximar coeciente de varianza

data1 = f.rvs( 5, 25, size=1000000)
data2 = f.rvs(15, 25, size=1000000)

ax=sns.distplot(data1, bins=500, kde=False, color='b')
ax=sns.distplot(data2, bins=500, kde=False, color='r')
ax.set(xlabel='Grados de libertad',ylabel='Frecuencia')
```

```python
# Calculo de probabilidad usando la distribucion f
f.pdf(4,15,25)
```

```python
# Calculo de probabilidad acumulada
f.cdf(4,15,25)
```

```python
f.ppf(0.9988900520089906,15,25)
```
