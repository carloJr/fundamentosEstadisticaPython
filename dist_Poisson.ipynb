```python
import matplotlib.pyplot as plt
%matplotlib inline

from scipy.stats import poisson
```

```python
lambda_p = 3

data = poisson.rvs(mu=lambda_p, size = 100)
len(data)
```

```python
mend, var, skew, kurt = poisson.stats(mu=lambda_p, moments= 'mvsk')
mend, var, skew, kurt
```

```python
import seaborn as sns
ax = sns.distplot(data, bins = 50, kde = False, color='green')
ax.set(xlabel ='Distribucion Binomial NEGATIVA', ylabel ='frecuencia')
ax.axvline(x=mend, linestyle='--',label='Media', c='r')

ax.legend()
```
