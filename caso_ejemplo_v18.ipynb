```python
# Definicion de HIPOTESIS
# En la media de una distribucion expontencial
from scipy.stats import expon
media, var, skew, kurt = expon.stats(scale = 30, moments='mvsk')
```

```python
# Paso 1: parametro lambda
# Paso 2: hipotesis parametro (HP)
mu = 30
mu > 30

# Paso 3:Mejor estimador
# Estimador
# paso 4:Distribucion 
promedio = np.mean(muestra)
promedio

desv = 2 
```

```python
# Paso 5 calculo del estimador y y su valor asociado estadistico
z = (promedio-mu)/np.sqrt(var/50)
z
```

```python
# Paso 6: estimado de rechazo y tolerancia al error 
alpha = 0.05 #tolerancia de error del 5%

# Criterio de rechazo
from scipy.stats import norm 
data_norm = norm.rvs(size=1000000)
```

```python
ax = sns.distplot(data_norm, bins=500,kde=False, color='blue')
ax.set_title('Distribucion normal')
```

```python
valor_critico = norm.ppf(1-alpha, loc = 0, scale =1)
valor_critico
```

```python
ax = sns.distplot(data_norm, bins=500,kde=False, color='blue')
ax.set_title('Distribucin normal')
ax.axvline(x=valor_critico, linestyle = '--',label='Valor critico',c='r')
ax.axvline(x=z, linestyle = '--',label='Valor estadistico',c='g')
ax.legend()
```
