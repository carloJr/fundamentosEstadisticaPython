```python
import matplotlib.pyplot as plt 
%matplotlib inline

from scipy.stats import expon
```

```python
data = expon.rvs(size=100000000)
len(data)
```

```python
mend, var, skew, kurt = expon.stats(moments= 'mvsk')
mend, var, skew, kurt
```

```python
import seaborn as sns
ax = sns.distplot(data, bins = 500, kde = True, color='green')
ax.set(xlabel ='Tiempo', ylabel ='Frecuencia')
ax.axvline(x=mend, linestyle='--',label='Media', c='r')
ax.set_title('Distribucion exponencial')

ax.legend()
```
