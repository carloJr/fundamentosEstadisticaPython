```python
import matplotlib.pyplot as plt 
%matplotlib inline

from scipy.stats import norm
```

```python
mead = 0
var = 1
data = norm.rvs(size =100000000)
len(data)
```

```python
mead, var, skew, kurt = norm.stats(moments='mvsk')
mead, var, skew, kurt
```

```python
import seaborn as sns 
ax = sns.distplot(data, bins=500,kde=True, color='green')
ax.set(xlabel='tiempo', ylabel='frecuencia')
ax.axvline(x = mead,linestyle='--',label='Media',c='r')
ax.axvline(x = var,linestyle='--',label='Varianza',c='b')
ax.set_title('Distribucion normal estandar')
ax.legend()
```
