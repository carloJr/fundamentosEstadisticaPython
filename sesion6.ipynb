```python
import scipy.stats
import numpy as np 
import pandas as pd 

import matplotlib.pyplot as plt 
import seaborn as sns 

%matplotlib inline
```

```python
df = pd.read_csv('bicicletas-compartidas.csv')
df.columns
```

```python
# Categoricas
y = df ['cuartil-ano'].apply(lambda x: 'cuartil-'+str(int(x)))
```

```python
fig, ax =plt.subplots()
ax.bar(y.value_counts().index, y.value_counts())
ax.set_xlabel('Cuatiles del a√±o')
ax.set_ylabel('Frecuencia')
```

```python
fig, ax =plt.subplots()
ax.bar(y.value_counts().index, y.value_counts())
ax.set_xlabel('Cuatiles del a√±o')
ax.set_ylabel('Frecuencia')
ax.patches[1].set_facecolor('red')
ax.patches[3].set_facecolor('yellow')
```

```python
 fig, ax = plt.subplots()
 ax.pie(y.value_counts(), labels = y.value_counts().index)
 ax.set_title('diagrama de pie')
```

```python
yv = df['viento']
```

```python
 fig, ax = plt.subplots()
 ax.hist(yv, bins = 30)
 ax.set_xlabel('Viento')
 ax.set_ylabel('Frecuencia')
 plt.axvline(np.mean(yv),c='r', linestyle ='--', label = 'Promedio')
 plt.axvline(np.mean(yv)+np.std(yv) ,c='k', linestyle ='--', label = '+ 1-Desviaci√≥n')
 plt.axvline(np.mean(yv)-np.std(yv) ,c='y', linestyle ='--', label = '- 1-Desviaci√≥n')
 ax.legend()
```
