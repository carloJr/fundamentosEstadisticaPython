```python
import pandas as pd # Analisis y manipulacion de datos
import numpy as np # vectores y matrices multidimensionales
import seaborn as sns # interfaz de alto nivel para graficos
```

```python
muestra = [42, 35, 29, 45, 41, 57, 54, 47, 48, 56, 47, 35, 52, 31, 52, 55, 57, 58, 26, 29, 32, 37, 32, 34, 48, 20, 48, 51, 27, 24, 39, 40, 31, 34, 23, 24, 41, 58, 44, 48, 31, 23, 27, 55, 43, 47, 30, 57, 38, 51]
```

```python
# Hipotesis
mu1 = 37
mu2 = 42
```

```python
 promedio = np.mean(muestra)
 promedio

```

```python
 desv = 2
```

```python
z_1 = (promedio-mu1)/desv
z_2 = (promedio-mu2)/desv
```

```python
from scipy.stats import norm
data1 = norm.rvs(loc = mu1, scale = desv, size= 1000000)
data2 = norm.rvs(loc = mu2, scale = desv, size= 1000000)
```

```python
ax = sns.distplot(data1, bins=500,kde=True,color='blue')
ax = sns.distplot(data2, bins=500,kde=True,color='red')
ax.axvline(x=promedio,c='g',linestyle='--',label='promedio muestral')
ax.legend()
```

```python
# Error tipo1: Probabilidad(p) de rechazar HO cuando este es verdadero
p_prom_mu1 =  norm.cdf(z_1)
1 - p_prom_mu1
```

```python
# Error tipo2: Probabilidad(p) de aceptar HO cuando este es falso
p_prom_mu2 =  norm.cdf(z_2)
p_prom_mu2
```
