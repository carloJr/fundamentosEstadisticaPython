```python
import matplotlib.pyplot as plt
%matplotlib inline

from scipy.stats import binom
```

```python
p = 0.3
n = 10

data = binom.rvs(p=p,n=n, size = 100)
len(data)
```

```python
mend, var, skew, kurt = binom.stats(p=p,n=n, moments= 'mvsk')
mend, var, skew, kurt
```

```python
import seaborn as sns
ax = sns.distplot(data, bins = 30, kde = False, color='green')
ax.set(xlabel ='Distribucion Binomial', ylabel ='frecuencia')
ax.axvline(x=mend, linestyle='--',label='Media', c='r')

ax.legend()
```
