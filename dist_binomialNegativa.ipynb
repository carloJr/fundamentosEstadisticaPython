```python
import matplotlib.pyplot as plt
%matplotlib inline

from scipy.stats import nbinom
```

```python
p = 0.3
n = 5

data = nbinom.rvs(p=p,n=n, size = 100000)
len(data)
```

```python
mend, var, skew, kurt = nbinom.stats(p=p,n=n, moments= 'mvsk')
mend, var, skew, kurt
```

```python
import seaborn as sns
ax = sns.distplot(data, bins = 50, kde = True, color='green')
ax.set(xlabel ='Distribucion Binomial NEGATIVA', ylabel ='frecuencia')
ax.axvline(x=mend, linestyle='--',label='Media', c='r')

ax.legend()
```
